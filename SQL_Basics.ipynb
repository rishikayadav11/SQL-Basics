**Q9.**  What is the Difference Between a Primary Key and a Foreign Key?

**Ans-** A **primary key** and a **foreign key** are both used in relational databases, but they serve different purposes.

| Primary Key                                                           | Foreign Key                                 |
| --------------------------------------------------------------------- | ------------------------------------------- |
| Uniquely identifies each record in a table.                           | Links one table to another.                 |
| Must contain unique values.                                           | Values can repeat.                          |
| Cannot be NULL.                                                       | Can be NULL (unless constrained otherwise). |
| There is only one primary key per table (though it can be composite). | A table can have multiple foreign keys.     |
| Ensures entity integrity.                                             | Ensures referential integrity.              |

**Example:**

 **Customers Table**

   | CustomerID (PK) | Name  |
   | --------------- | ----- |
   | 1               | Alice |
   | 2               | Bob   |

 **Orders Table**

   | OrderID (PK) | CustomerID (FK) | Amount |
   | ------------ | --------------- | ------ |
   | 101          | 1               | 50     |
   | 102          | 1               | 75     |
   | 103          | 2               | 30     |

- **CustomerID** in the **Customers table** is the **primary key** because it uniquely identifies each customer.

- **CustomerID** in the **Orders table** is a **foreign key** because it references the CustomerID primary key in the Customers table.



**Q10.**  What Are Constraints in SQL and Why Are They Used?

**Ans-** **CONSTRAINTS:** SQL constraints are rules applied to table columns to ensure that the data stored in the database is accurate, valid, and consistent. They help maintain data integrity by preventing invalid or incorrect data from being inserted, updated, or deleted.

**Common SQL Constraints**

 | Constraint      | Purpose                                                                        |
| --------------- | ------------------------------------------------------------------------------ |
| **NOT NULL**    | Ensures a column cannot contain NULL values.                                   |
| **UNIQUE**      | Ensures all values in a column are different.                                  |
| **PRIMARY KEY** | Uniquely identifies each record. It combines **NOT NULL** and **UNIQUE**.      |
| **FOREIGN KEY** | Creates a relationship between two tables and maintains referential integrity. |
| **CHECK**       | Ensures that values satisfy a specified condition.                             |
| **DEFAULT**     | Assigns a default value if no value is provided.                               |


**Example**

 CREATE TABLE Employees (

    EmployeeID INT PRIMARY KEY,
    Name VARCHAR(50) NOT NULL,
    Email VARCHAR(100) UNIQUE,
    Salary INT CHECK (Salary > 0),
    Department VARCHAR(50) DEFAULT 'HR'
 );

 **Why Are Constraints Used?**

- **Maintain data accuracy** by preventing invalid data.
- **Ensure data consistency** across the database.
- **Prevent duplicate records** using PRIMARY KEY and UNIQUE.
- **Maintain relationships between tables** using FOREIGN KEY.
- **Reduce data entry errors** by enforcing rules.
- **Improve overall database reliability and integrity**.
